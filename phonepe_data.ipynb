```python
#DATA TRANSFORMATION
#creating dataframe of aggreated transcations
path1 = "C:/Users/GVJai/Desktop/Project/PhonePe/Data/data/aggregated/transaction/country/india/state/"
columns1 = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],
            'Transaction_amount': []}

# Loop over the directories in the specified path
for state in os.listdir(path1):
    cur_state = os.path.join(path1, state)

# Loop over the directories within the state directory   
    for year in os.listdir(cur_state):
        cur_year = os.path.join(cur_state, year)

#Loop over the files within the year directory       
        for file in os.listdir(cur_year):
            cur_file = os.path.join(cur_year, file)

 # Open the file and load its content as JSON data           
            with open(cur_file, 'r') as data_file:
                A = json.load(data_file)

 # Loop over the 'transactionData' in the JSON data
                for i in A['data']['transactionData']:
                    name = i['name']
                    count = i['paymentInstruments'][0]['count']
                    amount = i['paymentInstruments'][0]['amount']
                    quarter = int(file.strip('.json'))
                    
# Append the extracted data to the respective lists in the columns dictionary                    
                    columns1['Transaction_type'].append(name)
                    columns1['Transaction_count'].append(count)
                    columns1['Transaction_amount'].append(amount)
                    columns1['State'].append(state)
                    columns1['Year'].append(year)
                    columns1['Quarter'].append(quarter)
# Create a DataFrame from the columns dictionary
df_agg_trans = pd.DataFrame(columns1)
```

```python
#creating dataframe of aggreated user
path2="C:/Users/GVJai/Desktop/Project/PhonePe/Data/data/aggregated/user/country/india/state/"
columns2 = {'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Count': [],
            'Percentage': []}
for state in os.listdir(path2):
    cur_state = os.path.join(path2, state)
    
    for year in os.listdir(cur_state):
        cur_year = os.path.join(cur_state, year)
        
        for file in os.listdir(cur_year):
            cur_file = os.path.join(cur_year, file)
            
            with open(cur_file, 'r') as data_file:
                B = json.load(data_file)
                try:
                   for i in B["data"]["usersByDevice"]:
                    brand_name = i["brand"]
                    counts = i["count"]
                    percents = i["percentage"]
                    quarter = int(file.strip('.json'))

                    columns2["Brands"].append(brand_name)
                    columns2["Count"].append(counts)
                    columns2["Percentage"].append(percents)
                    columns2["State"].append(state)
                    columns2["Year"].append(year)
                    columns2["Quarter"].append(quarter)
                except:
                 pass
df_agg_user = pd.DataFrame(columns2)
```

```python
#creating dataframe of map transcations
path3="C:/Users/GVJai/Desktop/Project/PhonePe/Data/data/map/transaction/hover/country/india/state/"

columns3 = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [],
            'Amount': []}

for state in os.listdir(path3):
    cur_state = os.path.join(path3, state)
    
    for year in os.listdir(cur_state):
        cur_year = os.path.join(cur_state, year)
        
        for file in os.listdir(cur_year):
            cur_file = os.path.join(cur_year, file)
            
            with open(cur_file, 'r') as data_file:
                C = json.load(data_file)
                for i in C["data"]["hoverDataList"]:
                    district = i["name"]
                    count = i["metric"][0]["count"]
                    amount = i["metric"][0]["amount"]
                    quarter = int(file.strip('.json'))
                    
                    columns3["District"].append(district)
                    columns3["Count"].append(count)
                    columns3["Amount"].append(amount)
                    columns3['State'].append(state)
                    columns3['Year'].append(year)
                    columns3['Quarter'].append(quarter)

df_map_trans = pd.DataFrame(columns3)
```

```python
#creating new tables
#creating agggrated transaction(Agg_trans) table
mysql_cursor.execute("CREATE TABLE IF NOT EXISTS agg_trans (State varchar(100), Year int, Quarter int, Transaction_type varchar(100), Transaction_count int, Transaction_amount double);")

for i,row in df_agg_trans.iterrows():
    #here %S means string values 
    sql = "INSERT INTO agg_trans VALUES (%s,%s,%s,%s,%s,%s)"
    mysql_cursor.execute(sql, tuple(row))
    # the connection is not auto committed by default, so we must commit to save our changes
    mysql_connection.commit()
```

```python
#creating agg_user table
mysql_cursor.execute("create table agg_user (State varchar(100), Year int, Quarter int, Brands varchar(100), Count int, Percentage double)")

for i,row in df_agg_user.iterrows():
    sql = "INSERT INTO agg_user VALUES (%s,%s,%s,%s,%s,%s)"
    mysql_cursor.execute(sql, tuple(row))
    mysql_connection.commit()
```
